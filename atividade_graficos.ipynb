```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
carros = pd.read_csv('used_cars_data.csv')
```

## 1- Distribuição das idades dos carros (anos de fabricação):

```python
carros['Year'] = pd.to_datetime(carros['Year'], format='%Y').dt.year

# Criar o histograma
plt.figure(figsize=(10, 6))
plt.hist(carros['Year'], bins=range(carros['Year'].min(), carros['Year'].max() + 1), color='skyblue', edgecolor='black')
plt.title('Distribuição das Idades dos Carros')
plt.xlabel('Ano de Fabricação')
plt.ylabel('Frequência')
plt.grid(True)
plt.xticks(range(carros['Year'].min(), carros['Year'].max() + 1), rotation=45)
plt.tight_layout()

# Mostrar o histograma
plt.show()
```

## 2- Distribuição dos preços dos carros:

```python
# Criar o histograma
plt.figure(figsize=(10, 6))
plt.hist(carros['Price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribuição dos Preços dos Carros')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.grid(True)
plt.tight_layout()

# Mostrar o histograma
plt.show()
```

## 3- Variação do preço em relação ao ano de fabricação:

```python
carros = pd.read_csv('used_cars_data.csv')

# Criar o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(carros['Year'], carros['Price'], alpha=0.5, color='blue')
plt.title('Variação do Preço em Relação ao Ano de Fabricação')
plt.xlabel('Ano de Fabricação')
plt.ylabel('Preço')
plt.grid(True)
plt.tight_layout()

# Mostrar o gráfico de dispersão
plt.show()
```

## 4- Tendência do preço dos carros ao longo dos anos:

```python


# Calcular a média dos preços para cada ano de fabricação
media_preco_por_ano = carros.groupby('Year')['Price'].mean().reset_index()

# Criar o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(media_preco_por_ano['Year'], media_preco_por_ano['Price'], marker='o', color='blue', linestyle='-')
plt.title('Tendência dos Preços dos Carros ao Longo dos Anos')
plt.xlabel('Ano de Fabricação')
plt.ylabel('Preço Médio')
plt.grid(True)
plt.tight_layout()

# Mostrar o gráfico de linha
plt.show()
```

## 5- Comparação da distribuição de preços entre diferentes tipos de locais:

```python

# Calcular a média dos preços para cada local de venda
media_preco_por_local = carros.groupby('Location')['Price'].mean().reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(media_preco_por_local['Location'], media_preco_por_local['Price'], color='skyblue')
plt.title('Comparação da Média de Preços entre Diferentes Locais')
plt.xlabel('Local de Venda')
plt.ylabel('Preço Médio')
plt.grid(True)
plt.tight_layout()

# Mostrar o gráfico de barras
plt.show()
```

## 6- Variação do preço em relação à quilometragem rodada:

```python
plt.figure(figsize=(10, 6))
plt.scatter(carros['Kilometers_Driven'], carros['Price'], alpha=0.5, color='blue')
plt.title('Variação do Preço em Relação à Quilometragem Rodada')
plt.xlabel('Quilometragem Rodada')
plt.ylabel('Preço')
plt.grid(True)
plt.tight_layout()

# Mostrar o gráfico de dispersão
plt.show()
```

## 7- Variação do preço em relação ao tipo de dono anterior:

```python
# Calcular a média dos preços para cada tipo de dono anterior
media_preco_por_dono = carros.groupby('Owner_Type')['Price'].mean().reset_index()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(media_preco_por_dono['Owner_Type'], media_preco_por_dono['Price'], color='skyblue')
plt.title('Comparação da Média de Preços por Tipo de Dono Anterior')
plt.xlabel('Tipo de Dono Anterior')
plt.ylabel('Preço Médio')
plt.grid(True)
plt.tight_layout()

# Mostrar o gráfico de barras
plt.show()
```
